```python
import io
import os
import glob
# Imports the Google Cloud client library
from google.cloud import vision
from google.cloud.vision import types
import pandas as pd

os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="/Users/damia/Desktop/Doggo_Chonkers/gcv_api.json"
client = vision.ImageAnnotatorClient()
```
